```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('build/lib.linux-x86_64-3.5')
%matplotlib inline
import beetracktracker
import pandas as pd
```

```python
mask  = cv2.imread('../data/BackWhite_mask.jpg',cv2.IMREAD_GRAYSCALE)
image = cv2.imread('../data/frame_00000000.jpg')
plt.imshow(mask)
```

```python
mybee=beetracktracker.beetracktracker()
mybee.set_mask(mask)
mybee.set_image(image)
predicted_centers=np.zeros((5,6))
mybee.set_predicted_centers(predicted_centers)

step=beetracktracker.beetracktracker.processing_steps.read
myim=mybee.get_data(step)
plt.imshow(myim)
```

```python
cap = cv2.VideoCapture("example/blackwhite_after_training.mp4")
frame_i=0;
ret, frame = cap.read()
import time;
time_managment=dict()
time_managment['read']=0
time_managment['set']=0
time_managment['run']=0
time_managment['get']=0
while(True):
    # Capture frame-by-frame
    start=time.time();
    ret, frame = cap.read()
    if ret==0:
        break
    time_managment['read']+=time.time()-start
    start=time.time();
    
    mybee.set_image(frame)
    time_managment['set']+=time.time()-start
    start=time.time();
    
    mybee.run()
    time_managment['run']+=time.time()-start
    start=time.time();
    
    step=beetracktracker.beetracktracker.processing_steps.contours_filtered
    myim=mybee.get_data(step)
    time_managment['get']+=time.time()-start
    start=time.time();
    
    frame_i+=1
    if frame_i>100:
        break
# When everything done, release the capture
cap.release()
```

```python
pd.DataFrame.from_dict(time_managment,orient='index')/frame_i
```
